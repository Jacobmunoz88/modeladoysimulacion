# Análisis Cualitativo de Sistemas de EDOs de Segundo Orden



# **Resumen**

El análisis cualitativo de vigas y columnas mediante ecuaciones diferenciales de segundo orden permite identificar los equilibrios y su estabilidad frente a cargas axiales, mostrando que la configuración recta constituye el equilibrio trivial estable mientras la carga aplicada sea menor que la carga crítica $P_{cr} = \frac{\pi^2 EI}{(KL)^2}$, la cual depende de las condiciones de apoyo; al alcanzar este valor, el sistema pierde estabilidad y aparecen equilibrios no triviales que corresponden a las formas de pandeo, de modo que para $P > P_{cr}$ cualquier pequeña imperfección se amplifica provocando la inestabilidad estructural.

# **Introduccion**
El estudio de la flexión y estabilidad de vigas y columnas constituye un problema clásico de la ingeniería estructural. Mediante la teoría de Euler–Bernoulli y el equilibrio de fuerzas se obtiene una ecuación diferencial ordinaria (EDO) de segundo orden que describe la forma de deformación lateral de la viga sometida a compresión axial. Este enfoque permite identificar los estados de equilibrio de la estructura y evaluar su estabilidad frente a perturbaciones.

# **Descripción del fenómeno**
El fenómeno de pandeo en vigas o columnas ocurre cuando una estructura delgada sometida a compresión axial, que inicialmente está en equilibrio recto, sufre una desviación lateral brusca al superar una carga crítica. Bajo cargas menores a dicho valor, la columna permanece estable y recupera su forma al retirar la carga; sin embargo, al alcanzar el nivel crítico, cualquier mínima imperfección geométrica o perturbación provoca que la deformación lateral crezca significativamente, dando lugar a una nueva configuración de equilibrio en forma curva conocida como modo de pandeo. Este fenómeno no se asocia a la falla por agotamiento del material, sino a una inestabilidad estructural dependiente de la esbeltez, rigidez a flexión y condiciones de apoyo de la viga o columna.

# **Modelo analítico y análisis cualitativo**

#Planteo general y definiciones

Consideramos una barra recta (viga/columna) de longitud $L$, módulo de Young $E$, momento de inercia de la sección $I$, sometida a una fuerza axial compresiva $P$ aplicada en el eje. Denotamos la desviación lateral (pequeña) por $y(x)$, con $x \in [0, L]$.

# Ecuación diferencial de pandeo (linealizada)

Para pequeñas desviaciones laterales se obtiene la ecuación (Euler — columna ideal, sin carga transversal):

$$ EI \frac{d^2y}{dx^2} + Py(x) = 0. \quad (1) $$

Nota importante: esta forma (segunda orden) es la que aparece al integrar la formulación de equilibrio de momentos bajo la hipótesis de pequeñas pendientes; es la forma clásica para obtener las cargas críticas y modos. (En presencia de cargas transversales distribuidas la ecuación completa de vigas es de orden cuatro:
$$ EI y''''(x) + Py''(x) = q(x). $$
Para el problema de pandeo sin carga transversal usamos la forma (1) como problema de autovalores espacial.)

Escribimos (1) como

$$ y'' + k^2 y = 0, \quad k^2 := \frac{P}{EI}. \quad (2) $$

Solución general:

$$ y(x) = A \cos(kx) + B \sin(kx). \quad (3) $$

Las constantes $A, B$ se determinan imponiendo las condiciones de contorno en $x=0$ y $x=L$. No triviales $(A, B) \neq (0, 0)$ sólo existen para valores discretos $k$ que satisfacen las condiciones; esos $k$ dan las autovalores $P = EI k^2$. El menor $P > 0$ no nulo es la carga crítica $P_{cr}$.

# Casos clásicos (condiciones de apoyo) — cálculo de $P_{cr}$ y modo fundamental

A continuación las condiciones y el resultado para el modo fundamental (menor $P$).

A) Apoyado–Apoyado (pinned–pinned)

Condiciones: $y(0)=0, y(L)=0$.
Imponer (3) da $A \cos 0 + B \sin 0 = A = 0 \Rightarrow A=0$. En $x=L$: $B \sin(kL) = 0 \rightarrow$ para $B \neq 0$ se requiere $\sin(kL)=0 \Rightarrow kL = n\pi$. Modo fundamental $n=1$.

$k_1 = \frac{\pi}{L} \Rightarrow P_{cr} = \frac{\pi^2 EI}{L^2}$.

Modo propio: $y(x) = B \sin \left( \frac{\pi x}{L} \right)$.

B) Empotrado–Empotrado (fixed–fixed)

Condiciones: $y(0)=0, y'(0)=0, y(L)=0, y'(L)=0$.
Imponer (3) produce la condición $\sin(kL)=0$ y adicionalmente $\cos(kL)=0$ para la misma combinación; en la práctica los $kL$ que satisfacen ambas dan $kL = n\pi$ pero el modo fundamental no es simplemente $n=1$ de la forma anterior por la restricción de pendiente; sin embargo, el resultado práctico suele expresarse mediante el factor de longitud efectiva (ver más abajo). El modo más rígido (fixed–fixed) tiene mayor $P_{cr}$; la forma propia típica tiene dos senos invertidos en el tramo.

En diseño se usa el factor de longitud efectiva para obtener $P_{cr}$ de forma uniforme (ver tabla abajo).

C) Empotrado–Libre (cantilever)

Condiciones: en $x=0$ empotramiento $y(0)=0, y'(0)=0$; en $x=L$ extremo libre $M(L)=0, V(L)=0$ equivalentes a $y''(L)=0, y'''(L)=0$ en formulación de cuarto orden. Para la ecuación simplificada (2) que usamos se obtiene la condición trascendental cuya primera raíz numérica da el modo fundamental con $kL \approx 1.87510407$. Por tanto

$P_{cr, cantilever} = EI \left( \frac{1.87510407}{L} \right)^2 \approx 3.516 \frac{EI}{L^2}$.

(Esta es la expresión correcta para el primer modo del cantilever en la formulación de columna.)

D) Empotrado–Apoyado (fixed–pinned)

La primera raíz de la condición mixta da $kL \approx 2.046$ y por tanto

$P_{cr, fixed-pinned} = EI \left( \frac{2.046}{L} \right)^2 \approx 4.186 \frac{EI}{L^2}$.

Tabla práctica (fórmula compacta con longitud efectiva)

Se usa con frecuencia la expresión

$P_{cr} = \frac{\pi^2 EI}{(KL)^2}$,

donde $K$ es el factor de longitud efectiva según condiciones de apoyo (valores típicos):

*   pinned–pinned (apoyado–apoyado): $K=1.0 \rightarrow P_{cr} = \pi^2 EI/L^2$.

*   fixed–fixed (empotrado–empotrado): $K \approx 0.5$.

*   fixed–pinned: $K \approx 0.7$ (más precisamente 0.699).

*   fixed–free (cantilever): $K=2.0$.

Usando la tabla: para cantilever $K=2 \Rightarrow P_{cr} = \pi^2 EI/(2L)^2 = \pi^2 EI/(4L^2) \approx 2.467 EI/L^2$.
Aclaración: la diferencia entre los valores “2.467” (fórmula con $K=2$) y el valor numérico $3.516 EI/L^2$ obtenido por la raíz $1.875^2$ viene de que existen dos modelos distintos (uno más exacto resuelto con la ecuación de orden 4 de Euler–Bernoulli y condiciones de momento/cortante en extremo libre). En ingeniería práctica se usan las tablas de $K$ (empírico/estándar) o se recurre a la forma completa con condiciones de contorno apropiadas. Para evitar confusión, el valor correcto teórico para cantilever (Euler–Bernoulli) es el que se obtiene resolviendo la ecuación completa y da $kL \approx 1.875$ con $P_{cr} \approx 3.516 EI/L^2$. El uso de la fórmula con $K$ es un atajo con valores tabulados según convención de diseño.


# Modos propios y forma crítica

Cada $P_{cr,n}$ está asociado a una forma propia $y_n(x)$ (modo de pandeo) con $n$ semiondas a lo largo de la columna.

El modo fundamental (n=1) es el que se alcanza primero al aumentar $P$ desde cero. Las formas propias determinan la geometría de la deflexión inicial cuando se pierde estabilidad.

#Interpretación de estabilidad (lineal)

Para $P < P_{cr}$: la solución $y \equiv 0$ es linealmente estable frente a perturbaciones laterales pequeñas (no crecerán). Si hay amortiguamiento dinámico cualquier pequeña oscilación decaerá.

Para $P = P_{cr}$: aparece un autovalor nulo en el problema lineal — punto de bifurcación: el equilibrio recto pierde estabilidad y aparecen soluciones laterales no triviales de arbitraria amplitud infinitesimal (en el análisis lineal).

Para $P > P_{cr}$: $y \equiv 0$ es linealmente inestable; pequeñas perturbaciones crecen (teóricamente sin límite en el modelo lineal). En la práctica la no linealidad geométrica/material limita la amplitud (post-pandeo).

Matemáticamente, pandeo es una bifurcación estática (para columnas simétricas suele ser una bifurcación tipo pitchfork).

#Efectos reales: imperfecciones y post-pandeo

Imperfecciones iniciales (pequeña curvatura inicial $y_0(x)$) hacen que no exista un cambio brusco en la carga: en la práctica la columna con imperfección adoptará una deflexión continua al aumentar $P$. Las imperfecciones reducen la carga efectiva de fallo respecto a $P_{cr}$ ideal.

Post-pandeo: la respuesta para $P > P_{cr}$ está gobernada por términos no lineales (geometría y material). Se puede obtener una rama estacionaria no nula (equilibrio lateral) que crece con $P$. El análisis requiere incluir términos no lineales (p. ej. efecto geométrico de la componente axial en la ecuación de equilibrio) o usar métodos de energía (Rayleigh–Ritz) para obtener la curva carga-deflexión.

Dinámica: si además consideras inercia y amortiguamiento, pequeñas perturbaciones para $P > P_{cr}$ crecerán exponencialmente en el tiempo (inestabilidad dinámica); hay fenómenos más complejos (pandeo dinámico, flutter) si hay carga axial variable o interacción con otras componentes (torsión).

#Método energético (criterio aproximado de $P_{cr}$)

Un método útil para obtener estimaciones aproximadas de $P_{cr}$ es usar una forma de energía potencial y un ensayo con una función aproximada $\varphi(x)$ (método de Rayleigh–Ritz). Esquema:

*   Suponer una forma de modo aproximada $y(x) = \eta \varphi(x)$ con amplitud $\eta$.

*   La energía potencial (cuando no hay carga transversal) es combinación de energía de curvado $U_b$ y del trabajo potencial de la carga axial $W_P$:

    $\Pi(\eta) = \frac{1}{2} EI \eta^2 \int_0^L (\varphi'')^2 dx - \frac{1}{2} P \eta^2 \int_0^L (\varphi')^2 dx$.

*   El primer valor de $P$ que hace que $\Pi$ deje de ser mínimo (es decir el prefactor cuadrático cambie de signo) da una aproximación de $P_{cr}$:

    $P_{cr (aprox)} = EI \frac{\int_0^L (\varphi'')^2 dx}{\int_0^L (\varphi')^2 dx}$.

(Si eliges $\varphi$ = forma real del primer modo, se obtiene el valor exacto; con funciones trial simples obtienes una cota superior o inferior.)

(Esta fórmula es útil para estimaciones rápidas y para vigas con condiciones o secciones complicadas donde obtener solución exacta es difícil.)

# Ejemplo numérico (cálculo de $P_{cr}$ para columna apoyada–apoyada)

Tomemos un ejemplo numérico sencillo:

$E = 210$ GPa, $I = 6.67 \times 10^{-6}$ m$^4$ (sección rectangular o circular equivalente), $L = 2.0$ m.

Entonces

$P_{cr} = \frac{\pi^2 EI}{L^2} = \frac{\pi^2 \cdot 210 \times 10^9 \cdot 6.67 \times 10^{-6}}{(2.0)^2}$.

Calculando numéricamente:

$P_{cr} \approx \frac{9.8696 \cdot 210 \times 10^9 \cdot 6.67 \times 10^{-6}}{4} \approx 3.45 \times 10^5 \text{ N} \approx 345 \text{ kN}$.

# **Modelo computacional y simulaciones**

```python
# Simulación PVI: dinámica lateral modal de una columna (simply-supported)
# Modo truncado: y(x,t) = sum_{n=1..N} q_n(t) * sin(n pi x / L)
# Integración temporal de q_n'' + (c_n/m_n) q_n' + (k_n/m_n) q_n = 0

import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# ---------------------------
# Parámetros físicos y numéricos
# ---------------------------
L = 1.0                     # longitud [m]
E = 2.1e11                  # módulo de Young [Pa]
I = 1e-10                   # inercia de la sección [m^4] (ajustable para escalado)
EI = E * I
rho = 7800.0                # densidad [kg/m^3] (ejemplo: acero)
A = 1e-4                    # área transversal [m^2]
mu = rho * A                # masa por unidad de longitud [kg/m]

# modos
N = 6                       # número de modos a retener
zeta = 0.01                 # "fracción" para construir c_n (damping ratio para cálculo de c_n)

# calcular P_cr (primer modo, pinned-pinned)
P_cr = (np.pi**2) * EI / (L**2)
print(f"P_cr (pinned-pinned, modo 1) = {P_cr:.3e} N")

# elegir dos casos: P < Pcr y P > Pcr
P_cases = {
    "P_bajo": 0.5 * P_cr,
    "P_alto": 1.2 * P_cr
}

# masa generalizada por modo (norma sin funciones ortonormales): integral sin^2 = L/2
m_n = mu * L / 2.0

# discretización espacial para reconstrucción
x_plot = np.linspace(0, L, 201)

# función para armar parámetros modales para una P dada
def build_modal_parameters(P):
    ns = np.arange(1, N+1)
    k_n = np.zeros(N)
    omega2 = np.zeros(N)
    c_n = np.zeros(N)
    for i, n in enumerate(ns):
        k_over_m = (EI * (n*np.pi/L)**4 - P * (n*np.pi/L)**2) / mu   # omega_n^2
        omega2[i] = k_over_m
        k_n[i] = m_n * k_over_m
        # amortiguamiento modal: c_n = 2 * zeta * sqrt(|k_n*m_n|)
        c_n[i] = 2.0 * zeta * np.sqrt(abs(k_n[i] * m_n))
    return k_n, c_n, omega2

# proyección de condición inicial y(x,0) sobre modos (fórmula para coeficentes q_n)
def project_initial_y(y_fun):
    # q_n(0) = (2/L) * integral_0^L y0(x) * sin(n pi x / L) dx
    q0 = np.zeros(N)
    for i, n in enumerate(range(1, N+1)):
        phi = lambda xx: y_fun(xx) * np.sin(n*np.pi*xx/L)
        # usar cuadratura simple (numpy trapz)
        xx = x_plot
        integrand = y_fun(xx) * np.sin(n*np.pi*xx/L)
        q0[i] = (2.0 / L) * np.trapz(integrand, xx)
    return q0

# RHS para el sistema modal (2N variables: q_1..q_N, v_1..v_N)
def modal_rhs(t, state, k_n, c_n):
    q = state[:N]
    v = state[N:]
    dqdt = v
    dvdt = np.zeros_like(v)
    for i in range(N):
        dvdt[i] = - (c_n[i] / m_n) * v[i] - (k_n[i] / m_n) * q[i]
    return np.concatenate([dqdt, dvdt])

# reconstrucción y evaluación de y(x,t) desde q(t)
def reconstruct_y(q_vec):
    yx = np.zeros_like(x_plot)
    for i, n in enumerate(range(1, N+1)):
        yx += q_vec[i] * np.sin(n*np.pi*x_plot/L)
    return yx

# ---------------------------
# Condición inicial: pequeña imperfección en primer modo
# ---------------------------
amp = 1e-4    # amplitud inicial de imperfección [m]
def y0_function(x):
    return amp * np.sin(np.pi*x/L)   # pura en modo 1 (puedes cambiar)

q0_init = project_initial_y(y0_function)
v0_init = np.zeros_like(q0_init)

# añadimos un pequeño ruido en modos superiores para excitar (opcional)
np.random.seed(1)
q0_init = q0_init + 1e-6 * np.random.randn(N)

# ---------------------------
# Simular cada caso (P bajo y P alto)
# ---------------------------
results = {}
t_final = 5.0   # tiempo final [s] (ajustar según tiempos naturales)
for name, P in P_cases.items():
    print(f"\nSimulando caso: {name} (P = {P:.3e} N)")
    k_n, c_n, omega2 = build_modal_parameters(P)
    # elegir t_final automático si frecuencias son altas: estimar periodo del primer modo si omega2 positive
    omega1_sq = omega2[0]
    if omega1_sq > 0:
        T1 = 2*np.pi / np.sqrt(omega1_sq)
        t_eval = np.linspace(0, max(5*T1, 0.5), 4000)
    else:
        # modo inestable -> escoger t span en base a tasa de crecimiento aproximada
        growth_rate = np.sqrt(max(0.0, -omega1_sq))
        t_eval = np.linspace(0, max(6.0/growth_rate if growth_rate>0 else 5.0, 0.5), 4000)
    # estado inicial
    state0 = np.concatenate([q0_init, v0_init])
    sol = solve_ivp(modal_rhs, (t_eval[0], t_eval[-1]), state0, t_eval=t_eval, args=(k_n, c_n), rtol=1e-8)
    results[name] = {
        "t": sol.t,
        "q": sol.y[:N,:],
        "v": sol.y[N:,:],
        "k_n": k_n,
        "c_n": c_n,
        "omega2": omega2,
        "P": P
    }

# ---------------------------
# Ploteo comparativo
# ---------------------------
plt.figure(figsize=(12,8))
for i, (name, data) in enumerate(results.items()):
    t = data["t"]
    q = data["q"]
    # reconstruir tip displacement y(L,t) = sum q_n(t) * sin(n*pi)
    sin_at_L = np.sin(np.arange(1, N+1) * np.pi)  # sin(n*pi)=0 for n integer -> but tip x=L yields 0 for pinned-pinned!
    # For simply supported the displacement is zero at x=L by BC, so show mid-span (x=L/2) instead:
    x_mid = L/2
    sin_at_mid = np.sin(np.arange(1, N+1) * np.pi * (x_mid/L))
    y_mid_t = q.T.dot(sin_at_mid)
    # primer modo q1
    q1 = q[0,:]
    plt.subplot(2,2,1 if i==0 else 2)
    plt.plot(t, y_mid_t, label=f"{name} (P={data['P']:.2e}N)")
    plt.xlabel('t [s]'); plt.ylabel('y(x=L/2,t) [m]')
    plt.title('Desplazamiento en medio del vano (x=L/2)')
    plt.grid(linestyle=':')
    plt.legend()
    # plot q1
    plt.subplot(2,2,3)
    plt.plot(t, q1, label=f"{name}")
    plt.xlabel('t [s]'); plt.ylabel('q1(t) [m]')
    plt.title('Amplitud del primer modo q1(t)')
    plt.grid(linestyle=':')
    plt.legend()

plt.tight_layout()
plt.show()

# ---------------------------
# Snapshots espaciales y fase (para cada caso)
# ---------------------------
for name, data in results.items():
    t = data["t"]
    q = data["q"]
    print(f"\nResultados para {name}: P = {data['P']:.3e} N")
    # elegir tiempos de snapshot: 4 tiempos equiespaciados
    snaps = np.linspace(0, t[-1], 5)
    plt.figure(figsize=(10,6))
    for ts in snaps:
        idx = np.argmin(np.abs(t - ts))
        qvec = q[:, idx]
        yx = reconstruct_y(qvec)
        plt.plot(x_plot, yx, label=f"t={t[idx]:.3f}s")
    plt.title(f"Snapshots de y(x,t) — {name}")
    plt.xlabel('x [m]'); plt.ylabel('y(x) [m]')
    plt.legend(); plt.grid(linestyle=':')
    plt.show()
    # mostrar escala log de q1 si crece
    q1 = q[0,:]
    plt.figure(figsize=(6,3))
    plt.plot(t, np.abs(q1), label='|q1(t)|')
    plt.yscale('log')
    plt.xlabel('t [s]'); plt.ylabel('|q1(t)| (escala log)')
    plt.title(f'Amplitud |q1(t)| (log) — {name}')
    plt.grid(linestyle=':')
    plt.show()

# ---------------------------
# Observaciones finales impresas
# ---------------------------
for name, data in results.items():
    omega2 = data['omega2']
    print(f"\n{name}: primeros omega2 (modos 1..3) = {omega2[:3]}")
    # indicar si el primer modo es estable/inestable
    if omega2[0] > 0:
        print("  Primer modo: omega1^2 > 0 -> comportamiento oscilatorio (estable si hay amortiguamiento).")
    elif abs(omega2[0]) < 1e-12:
        print("  Primer modo: casi crítico (omega1^2 ~ 0).")
    else:
        print("  Primer modo: omega1^2 < 0 -> inestable (crecimiento exponencial -> pandeo).")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp

# --------------------------
# Parámetros físicos / modales
# --------------------------
zeta = 0.05      # amortiguamiento
omega0 = 5.0     # frecuencia natural (rad/s) si P < Pcr
# Caso estable e inestable:
omega2_stable = omega0**2       # P < Pcr
omega2_unstable = - (omega0**2) # P > Pcr

# Sistema general: y'' + 2 zeta omega0 y' + omega2 y = 0
def system(t, Y, zeta, omega0, omega2):
    y, v = Y
    dy = v
    dv = -2*zeta*omega0*v - omega2*y
    return [dy, dv]

# --------------------------
# Simulación con diferentes condiciones iniciales
# --------------------------
t_span = (0, 10)
t_eval = np.linspace(0, 10, 1000)

# Condiciones iniciales distintas
ics = [
    (0.1, 0.0),   # pequeño desplazamiento, sin velocidad
    (0.0, 0.1),   # sin desplazamiento, con velocidad inicial
    (0.1, 0.2),   # desplazamiento + velocidad
]

cases = {
    "Estable (P < Pcr)": omega2_stable,
    "Inestable (P > Pcr)": omega2_unstable
}

# --------------------------
# Gráficas
# --------------------------
for case_name, omega2 in cases.items():
    plt.figure(figsize=(12,5))

    # Subplot 1: y(t) en el tiempo
    plt.subplot(1,2,1)
    for y0, v0 in ics:
        sol = solve_ivp(system, t_span, [y0,v0], t_eval=t_eval,
                        args=(zeta, omega0, omega2))
        plt.plot(sol.t, sol.y[0], label=f"y0={y0}, v0={v0}")
    plt.title(f"Trayectorias en el tiempo — {case_name}")
    plt.xlabel("t [s]"); plt.ylabel("y(t)")
    plt.legend(); plt.grid(True)

    # Subplot 2: espacio de fases
    plt.subplot(1,2,2)
    for y0, v0 in ics:
        sol = solve_ivp(system, t_span, [y0,v0], t_eval=t_eval,
                        args=(zeta, omega0, omega2))
        plt.plot(sol.y[0], sol.y[1], label=f"y0={y0}, v0={v0}")
    plt.title(f"Espacio de estados — {case_name}")
    plt.xlabel("y"); plt.ylabel("dy/dt")
    plt.legend(); plt.grid(True)

    plt.tight_layout()
    plt.show()
```

# **Análisis de resultados**

El análisis de los resultados obtenidos muestra que el comportamiento de las vigas o columnas sometidas a compresión axial depende directamente de la relación entre la carga aplicada y la carga crítica teórica de Euler. Para valores de $P < P_{cr}$, el equilibrio recto se mantiene estable y las deformaciones laterales son mínimas; al acercarse a $P_{cr}$, la estructura se vuelve muy sensible a pequeñas imperfecciones iniciales, incrementándose las deflexiones de forma notable; y para $P > P_{cr}$, la configuración recta pierde estabilidad y la columna adopta una forma de pandeo determinada por sus condiciones de apoyo. De este análisis se concluye que la estabilidad estructural no depende únicamente de la resistencia del material, sino también de la rigidez a flexión y la esbeltez de la viga, lo que hace necesario considerar factores de seguridad y métodos de diseño que contemplen las imperfecciones geométricas y las condiciones reales de apoyo para garantizar un desempeño seguro en estructuras comprimidas.

# **Conclusiones**

El análisis cualitativo basado en EDOs de segundo orden permite describir de manera sencilla el fenómeno de pandeo de columnas.

*   El equilibrio trivial (columna recta) es estable para cargas menores a $P_{cr}$, e inestable para cargas mayores.

*   El modo crítico depende directamente de las condiciones de apoyo y define la forma de la columna cuando pierde estabilidad.

*   En la práctica, imperfecciones geométricas y efectos no lineales reducen la carga real de pandeo respecto al valor teórico, por lo que se utilizan factores de seguridad en diseño estructural.
